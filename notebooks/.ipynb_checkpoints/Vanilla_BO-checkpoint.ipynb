```python
# Import some useful packages
from bayes_opt import BayesianOptimization
import numpy as np
from matplotlib import pyplot as plt
```

```python
'''Basic function I would never use for BO but might as well use as an example
   to get comfortable with these tools'''
def black_box_function(x):
    return -x ** 2
```

```python
# Restrict domain from -3 to 3

pbounds = {'x' : (-3,3)}
```

```python
# Declare the optimizer

optimizer = BayesianOptimization(
    f=black_box_function,
    pbounds=pbounds,
    verbose=2, # verbose = 1 prints only when a maximum is observed, verbose = 0 is silent
    random_state=1,
)
```

```python
# Minimize the optimizer with 5 iterations

optimizer.maximize(
    init_points=5,
    n_iter=5,
)
```

```python
(x_p,y_p) = optimizer.max.get("target"), optimizer.max.get("params").get("x")
```

```python
print(optimizer.max.get("params").get("x"))
```

```python
x = np.linspace(-10, 10)
```

```python
plt.plot(x, black_box_function(x), color='green')
plt.plot(x_p,y_p,'ro')
```
